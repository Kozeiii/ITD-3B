<h1> FILTERING DATA </h1>

```python
filter1 = terror[terror['region_txt'] == "Middle East & North Africa"]
```

```python
filter2 = filter1[filter1['iyear'] > 2000 ]
```

```python
filter2['attacktype1_txt'].value_counts()
```

```python
filter3 = filter2[filter2['attacktype1_txt'] == 'Bombing/Explosion']
```

```python
filter3
```

<h1> DATA AGGREGATION </h1>

```python
final_filter = filter3.groupby(['latitude', 'longitude']).size().reset_index(name = 'incident_count')
```

```python
final_filter
```

<h1> Use of FOLIUM FOR GEOSPATIAL </h1>

```python
import folium
from folium.plugins import HeatMap
```

<h1> CONVERTING Pandas Data frame into python list</h1>

```python
heatmap_data = final_filter[['latitude', 'longitude', 'incident_count']].values.tolist()
```

<h1> CREATING base map </h1>

```python
my_map = folium.Map(location = [30,40], zoom_start = 4)
```

<h1> Add Heatmap Layer</h1>

```python
HeatMap(heatmap_data).add_to(my_map)
```

```python
my_map.save('terrorism_bombing.html')
print("data saved")
```

```python
filtered_terror = terror[
    (terror['region_txt'] == 'South Asia') &
    (terror['country_txt'] == 'Pakistan') &
    (terror['iyear'] >= 2015) &
    (terror['attacktype1_txt'] == 'Bombing/Explosion')
].copy()

print(f"Filtered rows: {filtered_terror.shape[0]}")
```

<h1> NATIONALITY ANALYSIS </h1>

```python
filtered_nationality = terror[
    (terror['country_txt'] == 'Pakistan') &
    (terror['iyear'] >= 2015)
].copy()
```

```python
nationality_df = filtered_nationality[['natlty1_txt', 'latitude', 'longitude']].dropna()
map_natlty = folium.Map(location=[30, 70], zoom_start=4)
HeatMap(data=nationality_df[['latitude', 'longitude']]).add_to(map_natlty)

map_natlty
```

<h1> REGIONAL ANALYSIS </h1>

```python
new_region_df = terror[terror['region_txt'] == 'Sub-Saharan Africa'].copy()
map_new_region = folium.Map(location=[0, 20], zoom_start=3)
HeatMap(data=new_region_df[['latitude', 'longitude']].dropna()).add_to(map_new_region)

map_new_region
```

<h1> TARGET ANALYSIS </h1>

```python
filtered_target = terror[
    (terror['country_txt'] == 'Iraq') &
    (terror['iyear'].between(2010, 2012))
].copy()
```

```python
target_df = filtered_target[['targtype1_txt', 'latitude', 'longitude']].dropna()
map_target = folium.Map(location=[30, 70], zoom_start=4)
HeatMap(data=target_df[['latitude', 'longitude']]).add_to(map_target)

map_target
```

<h1> CASUALTY ANALYSIS </h1>

```python
filtered_casualty = terror[
    (terror['country_txt'] == 'Afghanistan') &
    (terror['iyear'] >= 2017)
].copy()
filtered_casualty['casualties'] = filtered_casualty[['nkill', 'nwound']].fillna(0).sum(axis=1)
```

```python
casualty_df = filtered_casualty[['casualties', 'latitude', 'longitude']].dropna()
map_casualties = folium.Map(location=[30, 70], zoom_start=4)
HeatMap(data=casualty_df[['latitude', 'longitude', 'casualties']].values.tolist()).add_to(map_casualties)

map_casualties
```

<h1> ATTACK TYPE ANALYSIS </h1>

```python
filtered_attack = terror[
    (terror['country_txt'] == 'India') &
    (terror['attacktype1_txt'] == 'Bombing/Explosion')
].copy()
```

```python
attack_df = filtered_attack[['attacktype1_txt', 'latitude', 'longitude']].dropna()

map_attack = folium.Map(location=[30, 70], zoom_start=4)
HeatMap(data=attack_df[['latitude', 'longitude']]).add_to(map_attack)

map_attack
```
